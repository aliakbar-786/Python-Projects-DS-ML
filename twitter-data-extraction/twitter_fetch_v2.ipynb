```python
!pip install tweepy

import tweepy  # Twitter API library
import json    # To handle JSON data
import datetime  # To handle timestamping output files

# === Step 1: Authentication ===

# These are Twitter API credentials — NEVER share them publicly.
# You should load these securely from environment variables or config files in production.
consumer_key = "vYvazxSr4gz19j2UNjQw8W6Fz"
consumer_secret = "QrEoXHQvz7bxKVrniSpEvC89GLDNk0Kua15LpIBYAVhTOtq6Qz"
access_key = "1557317528-L7MMHuKOnC73wbCBLpY9G6JwwounVSHn25aTDRB"
access_secret = "img3XUX9tShwOxBTunHs7wWmvisiTnK0QmUFnOyLHfODc"

# === Step 2: Configuration ===

# Define the keyword or hashtag to search tweets
search_query = "Pakistan"

# Get current timestamp and format it to use in file names (Windows doesn't allow ':' in names)
current_time = datetime.datetime.now().strftime('%Y-%m-%d-%H-%M')

# Set output filename using search query + timestamp
output_filename = f"{search_query}_{current_time}.json"

# === Step 3: Set Up Tweepy Authentication ===

# Authenticate using your Twitter API keys
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_key, access_secret)

# Create the Tweepy API object to access Twitter
api = tweepy.API(auth)

# === Step 4: Get Tweets Using Cursor ===

# Fetch 100 recent English-language tweets matching the search query
tweets = tweepy.Cursor(api.search_tweets,
                       q=search_query,
                       count=10,
                       lang="en",
                       tweet_mode='extended').items(100)

# === Step 5: Process and Store Tweets ===

tweets_list = []

# Loop through each tweet object
for tweet in tweets:
    tweets_list.append({
        "id": tweet.id,
        "id_str": tweet.id_str,
        "text": tweet.full_text,
        "source": tweet.source,
        "truncated": tweet.truncated,
        "user_tw": {
            "id": tweet.user.id,
            "id_str": tweet.user.id_str,
            "name": tweet.user.name,
            "screen_name": tweet.user.screen_name,
            "location": tweet.user.location,
            "url": tweet.user.url,
            "description": tweet.user.description,
            "translator_type": tweet.user.translator_type,
            "protected": tweet.user.protected,
            "verified": tweet.user.verified,
            "followers_count": tweet.user.followers_count,
            "friends_count": tweet.user.friends_count,
            "listed_count": tweet.user.listed_count,
            "favourites_count": tweet.user.favourites_count,
            "created_at": str(tweet.user.created_at),
            "utc_offset": tweet.user.utc_offset,
            "time_zone": tweet.user.time_zone
        }
    })

# === Step 6: Save Data to JSON File ===

with open(output_filename, 'w') as fout:
    json.dump(tweets_list, fout)

print(f"✅ {len(tweets_list)} tweets saved to {output_filename}")
```
